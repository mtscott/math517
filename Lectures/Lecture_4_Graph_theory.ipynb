# Creating Sparse Matrices in Python

Python's scipy.sparse module provides tools for working with sparse matrices. The most common formats are:

*   CSR (Compressed Sparse Row)

```python
import numpy as np
import networkx as nx
import scipy.sparse as sp
import matplotlib.pyplot as plt

# Create a sparse matrix
matrix = sp.csr_matrix([[1, 3, 0, 0,0], [3, 1, 4, 0,0], [0, 4, 1, 5,0], [0, 0, 5, 1,1],[0, 0, 0, 1,1]])

print(matrix)
# Convert the sparse matrix to a graph
graph = nx.from_scipy_sparse_array(matrix)
# Draw the graph
nx.draw(graph, with_labels=True)
print("Graph nodes", graph.nodes)
print("Graph edges", graph.edges)

plt.show()
```

```python
# Create the power of sparse matrix
matrix = matrix@matrix
# Convert the sparse matrix to a graph
graph = nx.from_scipy_sparse_array(matrix)
# Draw the graph
nx.draw(graph, with_labels=True)
print("Graph nodes", graph.nodes)
print("Graph edges", graph.edges)

plt.show()
```

```python
# generate two sparse matrices
matrix1 = sp.rand(5, 5, density=0.2, format='coo')
# Convert the sparse matrix to a graph
graph1 = nx.from_scipy_sparse_array(matrix1)
# Draw the graph
nx.draw(graph1, with_labels=True)
print("Graph 1 nodes", graph1.nodes)
print("Graph 1 edges", graph1.edges)



```

```python

# generate the second sparse matrix
matrix2 = sp.rand(5, 5, density=0.2, format='coo')
# Convert the sparse matrix to a graph
graph2 = nx.from_scipy_sparse_array(matrix2)
# Draw the graph
nx.draw(graph2, with_labels=True)
print("Graph 2 nodes", graph2.nodes)
print("Graph 2 edges", graph2.edges)

plt.show()
```

```python

# generate two sparse matrices
matrix3 = matrix1*matrix2 ## The graphs that are present in both (intersection)
# Convert the sparse matrix to a graph
graph3 = nx.from_scipy_sparse_array(matrix3)
# Draw the graph
nx.draw(graph3, with_labels=True)
print("Graph 3 nodes", graph3.nodes)
print("Graph 3 edges", graph3.edges)

plt.show()
```
